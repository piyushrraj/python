```python
import logging

logging.basicConfig(
    filename="check_sorted.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def is_sorted(lst: list) -> bool:
    """
    Check whether a list is sorted in ascending order.
    
    Parameters:
        lst (list): Input list of numbers
        
    Returns:
        bool: True if list is sorted in ascending order,
              otherwise False.
    """
    logging.info("Checking if the list is sorted in ascending order")

    for i in range(len(lst) - 1):
        if lst[i] > lst[i + 1]:
            logging.info("List is not sorted")
            return False

    logging.info("List is sorted")
    return True


numbers = [1, 2, 3, 4, 5]
result = is_sorted(numbers)

if result:
    print("The list is sorted in ascending order")
else:
    print("The list is not sorted")
```
