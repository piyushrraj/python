```python
def palindrome(num1):
    n1 = num1
    rev = 0
    while num1>0:
        ldigit = num1%10
        rev = rev*10 + ldigit
        num1 = num1//10
    return "Palindrome" if rev==n1 else "Not Palindrome"
print(palindrome(1211))
```

```python
def count_even_odd_digits():
    a = eval(input("Enter input: "))
    oddcount = 0
    evencount = 0
    while a != 0:
        digit = a % 10
        if digit % 2 == 0:
            evencount = evencount + 1
        else:
            oddcount = oddcount + 1
        a = a // 10
    return evencount, oddcount
even, odd = count_even_odd_digits()
print(even)
print(odd)
```
