```python
import logging

logging.basicConfig(
    filename="second_largest.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)
def second_largest(lst: list):
    """
    Find the second largest element in a given list
    Parameters:
        lst (list): Input list of numbers
    Returns:
        int : The second largest element in the list
    """
    logging.info("Finding second largest element in the list")
    if len(lst) < 2:
        logging.error("List must contain at least two elements")
        raise ValueError("List must contain at least two elements")
    largest = second = float('-inf')
    for num in lst:
        if num > largest:
            second = largest
            largest = num
        elif num > second and num != largest:
            second = num

    if second == float('-inf'):
        logging.error("No second largest element found (all elements may be equal)")
        raise ValueError("No second largest element found")

    logging.info(f"Second largest element calculated: {second}")
    return second
numbers = [10, 20, 4, 45, 99]
result = second_largest(numbers)
print("Second Largest Element:", result)
```
