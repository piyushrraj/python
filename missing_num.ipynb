```python
import logging

logging.basicConfig(
    filename="find_missing.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def find_missing(lst: list, n: int) -> int:
    """
    Find the missing number from a list containing numbers from 1 to n.
    
    Parameters:
        lst (list): List containing numbers from 1 to n with one missing
        n (int): Maximum number in the sequence
        
    Returns:
        int: Missing number
    """
    logging.info("Finding missing number in the list")

    total_sum = n * (n + 1) // 2
    list_sum = 0

    for num in lst:
        list_sum += num

    missing = total_sum - list_sum

    logging.info(f"Missing number calculated: {missing}")
    return missing


numbers = [1, 2, 4, 5]
result = find_missing(numbers, 5)
print("Missing Number:", result)
```
